```python
import numpy as np
import matplotlib.pyplot as plt

import maps
```

```python
resources = {}
icons = {}


def read_resource(filename: str):
    global resources
    if filename not in resources:
        img = plt.imread(filename)
        if img.shape[2] == 3:
            img = np.dstack((img, np.ones_like(img[:, :, 0])))

        resources[filename] = img

    return resources[filename]


def read_ground():
    img = read_resource("materials/terrains.png")
    return img[0: 32, :32]


def read_icon(filename: str, i: int):
    img = read_resource(filename)
    icon = img[i * 32: (i + 1) * 32, :32]
    alpha = np.expand_dims(icon[:, :, 3], 2)

    icon[:, :, :] = alpha * icon[:, :, :] + (1 - alpha) * read_ground()
    return icon
```

```python
def showmap(map: np.ndarray):
    global icons
    canvas = np.zeros((32 * 13, 32 * 13, 4))
    for i in range(13):
        for j in range(13):
            if map[i, j] not in icons:
                continue
            canvas[i * 32: (i + 1) * 32, j * 32: (j + 1)
                   * 32] = icons[map[i, j]]

    fig = plt.figure(figsize=(5, 5))
    ax = fig.add_subplot()
    ax.axis("off")
    ax.imshow(canvas)
    plt.show()
    plt.close()
```
